```python
import os
cwd = os.getcwd()
```

```python
import matplotlib.pyplot as plt
import csv

filename = cwd + r"\weights\csv_files\center0.5_height0.5_scale0.1.csv"
center = 0.5
centerHeight = 0.5
scale = 0.1

# Read data from CSV file
x = []
y = []
with open(filename, 'r') as file:
    reader = csv.reader(file)
    next(reader) # skip header
    for row in reader:
        x.append(int(row[0]))
        y.append(float(row[1]))

# Calculate percentage chance for center item
center_freq = y[x.index(int(center*100))]

# Create plot
fig, ax = plt.subplots()
ax.plot(x, y)

# Set plot title and axis labels
ax.set_title("Frequency of Items with 'Flattened' Scale Factor 0.1")
ax.set_ylabel("Frequency")

# Add legend showing minimum and maximum values and their difference
min_val = min(y)
max_val = max(y)
diff = max_val - min_val

# Add figtext showing percentage chance for center item
plt.figtext(0.5, 0, f"Percentage Chance for Center Item: {center_freq*100:.2f}%", ha="center")

plt.show()
```
