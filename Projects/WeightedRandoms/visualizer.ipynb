```python
import os
import matplotlib.pyplot as plt
import csv

def plot_data(filename, center, center_height, scale, plot_title):
    x = []
    y = []
    with open(filename, 'r') as file:
        reader = csv.reader(file)
        next(reader)  # skip header
        for row in reader:
            x.append(int(row[0]))
            y.append(float(row[1]))
    
    if center == 0:
        center = 0.01

    center_freq = y[x.index(int(center * 100))]

    fig, ax = plt.subplots()
    ax.plot(x, y)

    ax.set_title(plot_title)
    ax.set_ylabel("Frequency")

    min_val = min(y)
    max_val = max(y)
    diff = max_val - min_val

    plt.figtext(0.5, 0, f"Percentage Chance for Center Item: {center_freq * 100:.2f}%", ha="center")

    plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import re

# Function to extract center, height, and scale from the filename
def extract_params(filename):
    center = float(re.findall(r"center([0-9.]+)", filename)[0])
    height = float(re.findall(r"height([0-9.]+)", filename)[0])
    scale = float(re.findall(r"scale([0-9.]+)\.", filename)[0])
    return center, height, scale

# Load data from CSV files
data = {}
path = "Weights/csv_files"
file_list = [
    "center0.5_height0_scale1.csv",
    "center0.5_height0.5_scale1.csv",
    "center0.5_height0.95_scale1.csv",
]

for filename in file_list:
    center, height, scale = extract_params(filename)
    df = pd.read_csv(os.path.join(path, filename))
    data[(center, height, scale)] = df

# Plot the data
colors = ["red", "blue", "orange"]
fig, ax = plt.subplots()
for index, ((center, height, scale), df) in enumerate(data.items()):
    ax.plot(df["Item"], df["Frequency"], color=colors[index], label=f"Center: {center*100:.1f}%, Height: {height*100:.1f}%, Scale: {scale:.1f}")

# Customize plot
ax.legend(loc="upper center", bbox_to_anchor=(0.5, -0.15), ncol=1)
plt.xlabel("Item")
plt.ylabel("Frequency")
plt.title("Normal Distribution Centered with Varying Height")
plt.grid(True)
plt.tight_layout()
plt.show()
```
